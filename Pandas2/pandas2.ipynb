```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    # Read in data
    crime = pd.read_csv('crime_data.csv', index_col='Year')

    # (So we can see x and y labels.)
    # fig, ax = plt.subplots()

    # Plot 1: Histogram of Violent, Property, and Murder crimes.
    crime.plot(kind='hist', y=['Violent', 'Property', 'Murder'], ylabel="Crimes (in Millions)", stacked=True, alpha=0.7, bins=10, title="Occurance of Violent, Property, and Murder Crimes")

    # Plot 2: Line chart of Violent, Property, and Robery crimes.
    crime.plot(kind='line', y=['Violent', 'Property', 'Robbery'], xlabel='Occurances', title="Occurance of Violent, Property, and Robbery Crimes")
   
    # Plot 3: Line chart of population and total crimes.
    crime.plot(kind='line', y=["Population", 'Total'], title="Population and Total Crimes")
```

```python
prob1()
```

# Problem 2

```python
def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    # Read in the data
    crime = pd.read_csv('crime_data.csv')

    # Get crimes to compare forcible rape with
    compare_crimes = ["Violent", 'Burglary', 'Aggravated Assault']

    # Plot all the crimes against Forcible Rape
    for crimes in compare_crimes:
        crime.plot(kind='scatter', x='Forcible Rape', y=crimes, title=f"Forcible Rape and {crimes}", alpha=.88)

    # Return if correlated or not.
    return True, False, True
```

```python
prob2()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    # Read in data
    crimes = pd.read_csv('crime_data.csv')

    # Display distributions of Burglary, Violent, and Vehicle Theft
    crimes.plot(kind="box",xlabel='Occurances (in Millions)', y=['Burglary', 'Violent', 'Vehicle Theft'], vert=False, sharex=False, title="Theft Crimes")

    # Display distributions of Vehicle Thefts against the values of Robbery.
    crimes.plot(kind='hexbin', x='Vehicle Theft',xlabel="Vehicle Theft (in Millions)", y='Robbery', gridsize=25, title="Vehicle Theft Against Robbery")
```

```python
prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    e.g. what is the distribution of graduation rates or do schools 
    with lower student to faculty ratios have higher tuition costs.
    
    These plots should be easy to understand, 
    have clear variable names, and citations.
    """
    # Read in data
    college_info = pd.read_csv("college.csv", index_col="Name")
    
    # Graph s-t-f ratio against graduation rates.
    college_info.plot(kind="scatter", x="Grad.Rate", xlabel='Graduation Rate', y='S.F.Ratio', ylabel="Student to Faculty Ratio", \
                      title="Does Student-to-Faculty Ratio Affect Graduation Rate?", alpha=.8)
    
    # Graph private school against graduation rates.
    college_info["Private Grad Rate"] = college_info.query('Private == "No"')["Grad.Rate"]
    college_info["Non-Private Grad Rate"] = college_info.query('Private == "Yes"')["Grad.Rate"]
    college_info.plot(kind='box', y=['Non-Private Grad Rate', 'Private Grad Rate'], title="Graduation Rates at Non-Private and Private Schools", \
                      xlabel="Graduation Rates (in Percent)", vert=False)

    # Graph percentage of people who are accepted and actual enroll against how much it costs to attend (expend)
    college_info["Percent Attendance"] = 100*(college_info["Enroll"]/college_info["Accept"])
    college_info["Total Cost"] = college_info['Outstate'] + college_info["Room.Board"] + college_info["Books"] + college_info["Personal"]
    college_info.plot(kind='scatter', alpha=.8, x='Percent Attendance',xlabel='Percentage of Students Accepted who Actually Enroll (in Percent)', \
                      y='Total Cost', ylabel="How Much It Costs to Attend the School", title="Attendance Rates Against Price")
```
